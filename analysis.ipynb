```python
import requests
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# data connection
url = "http://127.0.0.1:5000/recipes" 

response = requests.get(url)
if response.status_code == 200:
    data = response.json()  # assuming the response is JSON data
else:
    print("Error: unable to retrieve data from endpoint")

df_recipes = pd.DataFrame(data)
```

```python
# recipe category barchart
fig1 = sns.countplot(x=df_recipes["category"])
fig1.set_xticklabels(fig1.get_xticklabels(),rotation=90)
print(fig1)
```

```python
# recipe preparation time bar chart
fig2 = sns.countplot(x=df_recipes["preparation_time"])
fig2.set_xticklabels(fig2.get_xticklabels(),rotation=90)
print(fig2)
```

```python
# recipe cooking time bar chart
fig3 = sns.countplot(x=df_recipes["cooking_time"])
fig3.set_xticklabels(fig3.get_xticklabels(),rotation=90)
print(fig3)
```

```python
# top 10 authors on bbcfood
top10authors = df_recipes['author'].value_counts()[:10]
fig4=sns.barplot(x=top10authors.index, y=top10authors.values)
fig4.set_xticklabels(fig4.get_xticklabels(), rotation=90)
fig4.set_title('Top 10 authors by number of recipes')
print(fig4)
```
